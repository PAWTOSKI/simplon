```python
# 11)Charger le fichier «pole_emploi.xlsx» après l’avoir ouvert et visualiser et l’assigner à la variable «p_emploi»

path='Téléchargements/dossier_simplon_chomage/pole_emploi.xlsx'
p_emploi=pd.read_excel(path,skiprows=3, columns=['code','libellé','P_E_2018'])
p_emploi.rename(columns={'Code':'REG','Libellé':'libellé', 'Pôle emploi 2018':'p_e_2018'}, inplace=True)
print(p_emploi)
```

```python
# 12) Joindre les data et p_emploiau seins d’un seul et unique dataframeen faisant correspondre les régions par leur codes. Utiliser pd.merge() pour réaliser cette jointure 

p_emploi.sort_values(by=['REG'])
data.sort_values('REG')
data_combo=p_emploi.merge(data, how='right')
print(data_combo[['REG','p_e_2018']].tail(50))
```

```python
#13) créer une colonne quantile dans p_emploi qui prendra les valeurs ['q25','q50','q75','q100'] en fonction du quartile auquel appartient le nombre de pôle emploi .pour se faire, utiliser "pd.qcut()"

liste_quantile=['q25','q50','q75','q100']
data_combo['quantile']=pd.qcut(data_combo['p_e_2018'],4,labels=liste_quantile )
data_combo['quantile'].head(50)
```

```python
14#)Générer un pivot de table montrant la moyenne du taux de chômage du trimestre 1 de l’année 2020 pour chaque régions et quantile du nombre de pôle emploi.

table=pd.pivot_table( data_combo,index=['LIBREG'],columns=['quantile'] ,values=['2020-T1'],aggfunc={'2020-T1':np.nanmean }  )   
print(table)
```

```python
#) expérimentation récréative
table2=pd.pivot_table( data_combo,index=['LIBREG'],columns=['quantile'] ,values=['2020-T1'],aggfunc={'2020-T1':np.sum }  )
print(table2)
```
